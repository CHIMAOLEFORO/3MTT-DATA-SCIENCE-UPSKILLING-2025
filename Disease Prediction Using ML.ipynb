## SPLITING OF TRAINING DATA BY ISOLATIONG THE TARGET COLUMN

## SPLITTING THE DATA USING TRAIN TEST  FUNCTION

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train.shape)
print(y_train.shape)
```

## MODEL BUILDING AND TRAINING

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd
from sklearn.preprocessing import StandardScaler
#from  sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
```

```python
### SCALING THE DATA
# Scale the data using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### FITTING A RANDOM FOREST MODEL

```python
# Initialize and train a Random Forest classifier
rf1 = RandomForestClassifier(n_estimators=100, random_state=42)
rf1.fit(X_train_scaled, y_train)
```

### MAKING PREDICTION

```python
# Make predictions on the test set
y_pred = rf1.predict(X_test_scaled)
```

### EVALUATING PERFORMANCE OF RANDOM FOREST CLASSIFIER

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

### USING A SUPPORT VECTOR MACHINE CLASSIFIER

```python
from sklearn.svm import SVC
import numpy as np
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.pipeline import make_pipeline
from  sklearn.metrics import classification_report,precision_score,recall_score,roc_auc_score
from sklearn.metrics import roc_auc_score
```

```python
##CREATING AN SVM PIPELINE
svm_pipeline1=make_pipeline(StandardScaler(),SVC(kernel='rbf', C=1.0,gamma='scale',probability=True,decision_function_shape='ovr'))
svm_pipeline1.fit(X_train_scaled, y_train)
```

```python
### CARRYING OUT PREDICTION
y_predSVM = svm_pipeline1.predict(X_test_scaled)
#predicting probabilities for ROC AUC
y_predSVM2 = svm_pipeline1.predict_proba(X_test_scaled)
```

### EVALUATING THE SVM PERFORMANCE

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_predSVM)
print("Accuracy of SVM:", accuracy)
print("Classification Report OF SVM:")
print(classification_report(y_test, y_predSVM))
print("Confusion Matrix of SVM:")
print(confusion_matrix(y_test, y_predSVM))
```
